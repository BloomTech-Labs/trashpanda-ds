```
from google_images_download import google_images_download

response = google_images_download.googleimagesdownload()
```

```
# using transparent keyword on arguments to get transparent background on all the images collected
arguments = {"keywords":"small appliances, batteries, tires","limit":20,"print_urls":True, "color_type": "transparent"} 
absolute_image_paths = response.download(arguments)
```

```
# using transparent and photo keywords on arguments to get transparent background on real images for all the images collected
arguments = {"keywords":"mixed paper","limit":20,"print_urls":True, "color_type": "transparent", "type": "photo"} 
absolute_image_paths = response.download(arguments)
```
